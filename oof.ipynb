```python
import pandas as pd
input_dir = "./backup/0210_141400/oof/"
oof0 = pd.read_csv(input_dir + "fold0_oof.csv")
oof1 = pd.read_csv(input_dir + "fold1_oof.csv")
oof2 = pd.read_csv(input_dir + "fold2_oof.csv")
oof3 = pd.read_csv(input_dir + "fold3_oof.csv")
oof4 = pd.read_csv(input_dir + "fold4_oof.csv")
tp_oof = pd.concat([oof0, oof1, oof2, oof3, oof4])
tp_oof
```

```python
oof0 = pd.read_csv(input_dir + "fold0_fp.csv")
oof1 = pd.read_csv(input_dir + "fold1_fp.csv")
oof2 = pd.read_csv(input_dir + "fold2_fp.csv")
oof3 = pd.read_csv(input_dir + "fold3_fp.csv")
oof4 = pd.read_csv(input_dir + "fold4_fp.csv")
fp_oof = pd.concat([oof0, oof1, oof2, oof3, oof4])
fp_oof = fp_oof.groupby(["recording_id", "patch"]).mean().reset_index()
fp_oof
```

```python
oof0 = pd.read_csv(input_dir + "fold0_test.csv")
oof1 = pd.read_csv(input_dir + "fold1_test.csv")
oof2 = pd.read_csv(input_dir + "fold2_test.csv")
oof3 = pd.read_csv(input_dir + "fold3_test.csv")
oof4 = pd.read_csv(input_dir + "fold4_test.csv")
test = pd.concat([oof0, oof1, oof2, oof3, oof4])
test = test.groupby(["recording_id", "patch"]).mean().reset_index()
test
```

```python
test.to_csv(input_dir + "test_kuto_v1.csv", index=False)
```
